```python


#df=pd.read_csv('csv\df_kl.csv')
df1=pd.read_csv(r'C:\Users\loges\Desktop\python\sample projects\GUVI\redbus\csv\df_ap.csv')
df2=pd.read_csv(r'C:\Users\loges\Desktop\python\sample projects\GUVI\redbus\csv\df_as.csv')
df3=pd.read_csv(r'C:\Users\loges\Desktop\python\sample projects\GUVI\redbus\csv\df_hp.csv')
df4=pd.read_csv(r'C:\Users\loges\Desktop\python\sample projects\GUVI\redbus\csv\df_kb.csv')
df5=pd.read_csv(r'C:\Users\loges\Desktop\python\sample projects\GUVI\redbus\csv\df_rs.csv')
df6=pd.read_csv(r'C:\Users\loges\Desktop\python\sample projects\GUVI\redbus\csv\df_sb.csv')
df7=pd.read_csv(r'C:\Users\loges\Desktop\python\sample projects\GUVI\redbus\csv\df_tl.csv')
df8=pd.read_csv(r'C:\Users\loges\Desktop\python\sample projects\GUVI\redbus\csv\df_up.csv')
df9=pd.read_csv(r'C:\Users\loges\Desktop\python\sample projects\GUVI\redbus\csv\df_wb.csv')

df_names = ['bus_details_ap','bus_details_as','bus_details_hp','bus_details_kb','bus_details_rs','bus_details_sb','bus_details_tl','bus_details_up','bus_details_wb']
dfs = [df1, df2, df3, df4, df5, df6, df7, df8, df9]

# Assuming df is a pandas DataFrame with columns 'RouteName' and 'Links'
# df = pd.DataFrame({'RouteName': [...], 'Links': [...]})
def scrape_bus_data(df,df_name:str):
    driver = webdriver.Chrome()

    bus_name = []
    bust_ype = []
    start_time = []
    end_time = []
    rating = []
    tota_lduration = []
    price = []
    seats_available = []
    route_name = []
    route_link = []

    for i, r in df.iterrows():
        routes = r['RouteNmae']
        links = r['Links']
        
        driver.get(links)
        time.sleep(2)

        # Create elements for looping by XPath
        elements = driver.find_elements(By.XPATH, f'//a[contains(@href, "{links}")]')
    
        # Looping through elements
        for element in elements:
            element.click()
            time.sleep(2)
            
        try:
            clicks = driver.find_elements(By.XPATH, '//div[@class="button"]')
            for click in clicks:
                click.click()
        except Exception as e:
            print("Error clicking button:", e)
            continue
            
        time.sleep(2)

        scrolling = True
        while scrolling:
            old_page_source = driver.page_source

            # Use action chains to perform page down
            actions = ActionChains(driver)
            actions.key_down(Keys.PAGE_DOWN).perform()
            time.sleep(5)

            new_page_source = driver.page_source
            if old_page_source == new_page_source:
                scrolling = False

        # Data extraction
        bus_name = driver.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
        bus_type = driver.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
        start_time = driver.find_elements(By.XPATH, "//div[@class='dp-time f-19 d-color lh-24']")
        end_time = driver.find_elements(By.XPATH, "//div[@class='bp-time f-19 d-color lh-24']")
        try:
            ratings = driver.find_elements(By.XPATH, "//div[@class='clearfix row-one']/div[@class='column-six p-right-10 w-10 fl']")
        except:
            ratings = []
        price = driver.find_elements(By.XPATH, '//div[@class="fare d-block"]')
        total_duration = driver.find_elements(By.XPATH, '//div[@class="dur l-color lh-24"]')
        seats_available = driver.find_elements(By.XPATH, '//div[contains(@class, "seat-left m=top-30")]')

        # Append data to respective lists
        for bus in bus_name:
            bus_name.append(bus)
            route_link.append(links)
            route_name.append(routes)
        
        for bus_type_ in bus_type:
            bus_type.append(bus_type_.text)
        
        for start_time_ in start_time:
            start_time.append(start_time_.text)

        for end_time_ in end_time:
            end_time.append(end_time_.text)

        for rating_ in ratings:
            rating.append(rating_.text)

        for price_ in price:
            price.append(price_.text)

        for totalduration_ in total_duration:
            total_duration.append(totalduration_.text)

        for seat in seats_available:
            seats_available.append(seat.text)

    # Creating a DataFrame to store the data
    data = {
        'BusName': bus_name,
        'BusType': bus_type,
        'StartTime': start_time,
        'EndTime': end_time,
        'Rating': rating,
        'Price': price,
        'TotalDuration': total_duration,
        'SeatsAvailable': seats_available,
        'RouteName': route_name,
        'Links' : route_link
    }
    df_name= pd.DataFrame(data)
    df_name.to_csv(f'csv/{df_name}.csv', index=False)
    driver.quit()
```
